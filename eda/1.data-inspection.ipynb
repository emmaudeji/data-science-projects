```python
product_analysis = (
    df.groupby("Item Purchased")
      .agg(
          Sales=("Item Purchased","count"),
          Revenue=("Purchase Amount (USD)","sum")
      )
      .sort_values(
          "Revenue",
          ascending=False
      )
)

product_analysis.head(15)
```

```python
product_analysis.columns
```

```python
sales_median = product_analysis["Sales"].median()
revenue_median = product_analysis["Revenue"].median()
```

```python
def classify(row):

    if row["Sales"] >= sales_median and row["Revenue"] >= revenue_median:
        return "Star"

    elif row["Sales"] >= sales_median:
        return "Popular"

    elif row["Revenue"] >= revenue_median:
        return "Premium"

    else:
        return "Weak"

product_analysis["Segment"] = (
    product_analysis.apply(
        classify,
        axis=1
    )
)
```

```python
product_analysis.groupby(
    "Segment"
).size()
```
