```python
df.groupby(
"Subscription Status"
)["Purchase Amount (USD)"]\
.mean()
# sns.boxplot(
# data=df,
# x="Subscription Status",
# y="Purchase Amount (USD)"
# )
```

```python
subscription_spend = (
    df.groupby("Subscription Status")
      ["Purchase Amount (USD)"]
      .mean()
)

non_subscribers = subscription_spend["No"]
subscribers = subscription_spend["Yes"]

percentage_difference = (
    (subscribers - non_subscribers)
    / non_subscribers
) * 100

print(percentage_difference)
```
